```python
train_path = next(os.walk('Atrophymasked/'))[2]
train_ids = sorted(train_path)
# train_ids = train_ids[317:]
# train_ids = ['P0166.jpg','P0200.jpg','P0130.jpg']
print(train_ids)
```

```python
path = 'PPA_images'
entr=[]
countpix=[]
count=1
labels=[]
import math
def truncate(number, digits) -> float:
    stepper = 10.0 ** digits
    return math.trunc(stepper * number) / stepper
for filename in train_ids:
    mId = filename.split('.')[0]
    img = cv2.imread('Atrophymasked/'+mId+'.jpg',0)
    img=cv2.resize(img,(128,128))
    #img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    if count < 18:
        namee = 'HighlyMyopic'
        labels.append(namee)
    elif count < 102:
        namee = 'Normal'
        labels.append(namee)
    elif count < 312:
        namee = 'PathologicallyMyopic'
        labels.append(namee)
    count=count+1
    glcm = np.squeeze(greycomatrix(img, distances=[1], 
                               angles=[0,45,90,135,180,225,270,315,360], symmetric=True, normed=True))
    entropy = -np.sum(glcm*np.log2(glcm + (glcm==0)))
    # yields 10.704625483788325
    print(entropy)
    entropy=truncate(entropy,5)
    entr.append(entropy)


    bright_count = np.sum(np.array(img) < 255)
    print(bright_count)
    countpix.append(bright_count)
    
f = open("Posterior1.csv", "w")
#f.write("{},{},{}\n".format("Entropy","Pix_Count","class"))
for x in zip(entr,countpix,labels):
    f.write("{},{},{}\n".format(x[0], x[1], x[2]))
cv2.waitKey()
cv2.destroyAllWindows()
```
